## Step 0: Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 1: Loading the Data

### Task 1.1: Loading the IMDb Dataset

```python
imbd = pd.read_csv("imbd.csv" , sep = ',')
imbd
```

### Task 1.2: Loading the Movies Dataset

```python
movies = pd.read_csv("movies.csv" , sep = ',')
movies
```

### Task 1.3: Merging the Datasets

```python
merged_data = pd.merge(imbd , movies)
```

### Task 4.2: Creating a Scatter Plot of Movie Runtime vs. IMDb Rating

```python
plt.scatter(merged_data['Runtime'], merged_data['IMDB_Rating'], alpha=0.7)
plt.title('Movie Runtime VS. IMDb Rating')
plt.xlabel('RunTime')
plt.ylabel('IMDB Rating')
plt.show()
```

## Step 5: Identifying the Top-Rated Directors in Each Genre

### Task 5.1: Extracting the Top Director for Each Genre

```python
topdir = merged_data.groupby(['Genre', 'Director'])['IMDB_Rating'].mean().reset_index()
topdir = topdir.groupby('Genre').first()
topdir = topdir.sort_values(by= 'IMDB_Rating' , ascending=False)
topdir
```

### Task 5.2: Counting the Occurrences of Top Directors by Genre

```python
topdir_counts = topdir['Director'].value_counts().reset_index()
```

## Step 7: Exploring Additional Questions

### Question 1: How does the movie certificate (G, PG, PG-13, R, etc.) impact the IMDb ratings?

```python
certificate = merged_data.dropna(subset=['Certificate', 'IMDB_Rating'])
certificate_to_rating = merged_data.groupby('Certificate')['IMDB_Rating'].mean().reset_index()
certificate_to_rating = certificate_to_rating.sort_values(by='IMDB_Rating', ascending=False)
plt.figure(figsize=(14, 5))
plt.bar(certificate_to_rating['Certificate'], certificate_to_rating['IMDB_Rating'])
plt.title('Average IMDb Rating by Movie Certificate')
plt.xlabel('Certificate')
plt.ylabel('Average IMDb Rating')
plt.show()
```

### Question 2: What is the relationship between the Metascore and IMDb ratings of movies?

```python
merged_data = merged_data.dropna(subset=['Meta_score', 'IMDB_Rating'])
metascore_to_rating = merged_data.groupby('Meta_score')['IMDB_Rating'].mean().reset_index()
metascore_to_rating = metascore_to_rating.sort_values(by='IMDB_Rating', ascending=False)
plt.figure(figsize=(14, 5))
plt.bar(metascore_to_rating['Meta_score'], metascore_to_rating['IMDB_Rating'])
plt.title('Average IMDb Rating by Movie Certificate')
plt.xlabel('Meta_score')
plt.ylabel('Average IMDb Rating')
plt.show()
```

### Question 3: Are there any trends in the gross earnings of movies across different genres?

```python
merged_data = merged_data.dropna(subset=['Gross', 'Genre'])

merged_data['Gross'] = merged_data['Gross'].astype(str)

# Clean 'Gross' column by removing non-numeric characters and converting to numeric
merged_data['Gross'] = merged_data['Gross'].str.replace(',', '').astype(float)

# Calculate the average gross earnings by genre
genre_gross = merged_data.groupby('Genre')['Gross'].mean().reset_index()

# Sort the result by the average gross earnings in descending order
genre_gross = genre_gross.sort_values(by='Gross', ascending=False)

# Plot the average gross earnings by genre
plt.figure(figsize=(55, 20))  # Adjust figure size as needed
plt.bar(genre_gross['Genre'], genre_gross['Gross'], color='skyblue')
plt.title('Average Gross Earnings by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Gross Earnings (in millions)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
